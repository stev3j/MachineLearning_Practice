```python
fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
fish.head()
```

```python
# fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_input
```

```python
# fish_target = fish['Species'].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
fish_input
```

```python
pd.unique(fish['Species'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
train_input
```

# 다중 분류 ( k - 최근접 이웃 )

> 3개 이상의 레이블로 분류



```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_input, train_target)
```

```python
print(kn.classes_) # 타겟 종류?
```

```python
kn.score(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
distance, indexes = kn.kneighbors(test_input[3:4])
print(train_target[indexes])
```

```python
distance, indexes = kn.kneighbors(test_input[2:3])
print(train_target[indexes])
```

```python
distance, indexes = kn.kneighbors(test_input[0:5])
print(train_target[indexes])
```

```python
bream_smelt_indexes  = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_input[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
```

```python
lr.predict(train_bream_smelt[:5])
```

```python
print(lr.coef_, lr.intercept_)
```

```python
decisions = lr.decision_function(train_bream_smelt[:5])
```
