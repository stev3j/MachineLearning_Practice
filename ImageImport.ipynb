```python
import gdown
import pandas as pd
import numpy as np
```

```python
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
ns_book7 = pd.read_csv('ns_book7.csv')
ns_book7
```

```python
count_by_year = ns_book7['발행년도'].value_counts() # 발행년도의 값을 리스트로 갖고옴
```

```python
count_by_year
```

```python
count_by_year = count_by_year.sort_index() # 정렬
```

```python
count_by_year
```

```python
count_by_year = count_by_year[count_by_year.index <= 2030] # 2030보다 높은 년도를 없앰 (2650책은 없잖아..)
```

```python
count_by_year
```

```python
def kdc_lst_char(no):
  if no is np.nan:
    return '-1'
  else:
    return no
```

```python
count_by_subject = ns_book7['주제분류번호'].apply(kdc_lst_char).value_counts()
```

```python
count_by_subject
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
```

```python
plt.plot(count_by_year.index, count_by_year.values) # x : 년도 , y : 년도에 따른 책의 양
plt.title('Boks by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, marker='.', color='red')
plt.title('Boks by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, '*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))
for idx, val in count_by_year[::5].items():
  plt.annotate(val, (idx, val), xytext=(idx+1, val+10))
plt.savefig('books_by_year')
plt.show()
```
